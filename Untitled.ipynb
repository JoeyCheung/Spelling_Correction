```python
# function to read raw text file
def read_text(filename):
    # open the file
    file = open(filename, mode='rt', encoding='utf-8')
    # read all text
    text = file.read()
    file.close()
    return text
```

```python
# split a text into sentences
def to_lines(text):
    sents = text.strip().split('\n')
    sents = [i.lower() for i in sents]
    return sents
```
